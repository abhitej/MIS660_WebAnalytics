```python
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
df = pd.read_csv('reviews.txt',delimiter='\t', header =None)
df
```

```python
df_clean = pd.DataFrame()

df_clean['Tomatometer'] = df[0].str.replace(r"[a-zA-Z:% ]",'')
df_clean['Average_Rating'] = df[1].str.replace(r"[a-zA-Z:% ]",'')
df_clean['Reviews_Counted'] = df[2].str.replace(r"[a-zA-Z:% ]",'')
df_clean['Fresh'] = df[3].str.replace(r"[a-zA-Z:% ]",'')
df_clean['Rotten'] = df[4].str.replace(r"[a-zA-Z:% ]",'')
df_clean['Audience_Rating'] = df[5].str.replace(r"[a-zA-Z:% ]",'')
df_clean['A_Average_Rating'] = df[6].str.replace(r"[a-zA-Z:% ]",'')
df_clean['A_User_Rating'] = df[7].str.replace(r"[a-zA-Z:% ]",'')
```

```python
df_clean['Average_Rating'] = df_clean.Average_Rating.str.replace('/10', '')
df_clean['A_Average_Rating'] = df_clean.A_Average_Rating.str.replace('/5', '')

df_clean.dtypes
```

```python
# convert empty strings to numpy to use numpy features

df_clean['Tomatometer'].replace('', np.nan, inplace=True)
df_clean['Average_Rating'].replace('', np.nan, inplace=True)
df_clean['Reviews_Counted'].replace('', np.nan, inplace=True)
df_clean['Fresh'].replace('', np.nan, inplace=True)
df_clean['Rotten'].replace('', np.nan, inplace=True)
df_clean['Audience_Rating'].replace('', np.nan, inplace=True)
df_clean['A_Average_Rating'].replace('', np.nan, inplace=True)
df_clean['A_User_Rating'].replace('', np.nan, inplace=True)
```

```python
sb.heatmap(df_clean.isnull())  # check for nan
```

```python
df_clean.dropna(inplace=True)
```

```python
sb.heatmap(df_clean.isnull()) 
```

```python
# convert to Float

df_clean[['Tomatometer', 'Average_Rating', 'Reviews_Counted', 'Fresh', 'Rotten', 'Audience_Rating',
          'A_Average_Rating','A_User_Rating']] = df_clean[['Tomatometer', 'Average_Rating', 'Reviews_Counted',
          'Fresh', 'Rotten', 'Audience_Rating','A_Average_Rating','A_User_Rating']].astype(float)

df_clean.dtypes
```
